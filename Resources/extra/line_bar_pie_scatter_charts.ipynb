### LINE GRAPHS

```python
%matplotlib inline
```

**NOTE**: Using inline helps make sure our plots are displayed inline within the Jupyter Notebook directly below the code cell that produced it.

```python
# import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()  Second Way of Writing it
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot with annotations and color changes
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.xlim()
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid(color='grey', linewidth='0.5')
# Add the legend.
plt.legend()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis, marker='D',color='green',linewidth=2)
plt.ylim(0,45)
plt.xlabel('DATE')
plt.ylabel('FARE ($)')
plt.title('PYBER FARE BY MONTH')
plt.legend('Boston')
ax.grid()
```

### PIE GRAPHS

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# matlab approach
# plt.subplots(figsize=(8, 8))
plt.pie(y_axis, labels=x_axis)
# explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
# plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# adding explode value
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))
colors = ["slateblue", "magenta", "lightblue", "green", 
          "yellowgreen", "greenyellow", "yellow", "orange", 
          "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis,
       explode=explode_values,
       colors=colors,
       shadow=True,
       startangle=90,
       counterclock=False,
       labels=x_axis,
       autopct='%.1f%%')

plt.show()
```

### SCATTER PLOTS

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# matlab method
plt.plot(x_axis, y_axis, 'o')
```

```python
# plt.scatter() function
plt.scatter(x_axis, y_axis)
```

```python
# skill drill using matlab plotting approach
plt.plot(y_axis, x_axis, 'o', color='red')
plt.gca().invert_yaxis()
plt.legend('CHICAGO')
plt.xlim(0,45)
plt.xlabel('DATE')
plt.ylabel('FARE ($)')
plt.title('PYBER FARE BY MONTH')
```

```python
# how to make the dots larger
plt.scatter(x_axis, y_axis, s= [i*3 for i in y_axis])
# y_axis_larger = []
# for data in y_axis:
#     y_axis_larger.append(data*3)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s= [i*5 for i in y_axis],color='skyblue',linewidth=2,alpha=.2,edgecolor='black')
plt.gca().invert_yaxis()
plt.xlim(0,50)
plt.title('PYBER FARE BY MONTH')
ax.legend('Chicago')
plt.xlabel('MONTHS')
plt.ylabel('FARE ($)')
```

**NOTE**: Adding error bars can show either the standard deviation, standard error, confidence intervals, or minimum and maximum values of a dataset. When added to a chart, they can visually show the variability of the plotted data. By looking at the error bars, one can infer the significance of the data.

To create a line chart with error bars using the MATLAB approach, we use the errorbar() function. We also add the x and y data parameters and the yerr= parameter inside parentheses, as shown here:
