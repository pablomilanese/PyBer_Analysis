```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# matlab method
plt.plot(x_axis, y_axis, 'o')
```

```python
# plt.scatter() function
plt.scatter(x_axis, y_axis)
```

```python
# skill drill using matlab plotting approach
plt.plot(y_axis, x_axis, 'o', color='red')
plt.gca().invert_yaxis()
plt.legend('CHICAGO')
plt.xlim(0,45)
plt.xlabel('DATE')
plt.ylabel('FARE ($)')
plt.title('PYBER FARE BY MONTH')
```

```python
# how to make the dots larger
plt.scatter(x_axis, y_axis, s= [i*3 for i in y_axis])
# y_axis_larger = []
# for data in y_axis:
#     y_axis_larger.append(data*3)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s= [i*5 for i in y_axis],color='skyblue',linewidth=2,alpha=.2,edgecolor='black')
plt.gca().invert_yaxis()
plt.xlim(0,50)
plt.title('PYBER FARE BY MONTH')
ax.legend('Chicago')
plt.xlabel('MONTHS')
plt.ylabel('FARE ($)')
```
